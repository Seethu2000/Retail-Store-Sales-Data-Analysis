**Phase 2 – Data Pre-processing**

**1.Handling Missing Values**

```python
# check if there is missing values
df.isnull().sum()
```



* **Item Column**

The Item column represents the product name involved in each transaction. During the initial data exploration, it was observed that this column contains 1213 missing values. Since the Item column is a categorical variable, missing values were handled by replacing them with the most frequent value (mode) in the column.



```python
# fill missing values in item column using mode
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
```



* **Price Per Unit Column**

The Price Per Unit column contains 609 missing values. Since this column is a numerical variable, the missing values are filled using the median value. The median is used instead of the mean because it is less affected by extreme values (outliers) and provides a more reliable central value for the dataset.



```python
# fill missing price values using median
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
```



* **Quantity column**

The Quantity column represents the number of items purchased in each transaction. During data exploration, it was found that this column contains 604 missing values. Since Quantity is a numerical column, the missing values are filled using the median value. The median is used because it is less affected by extreme values and gives a more reliable central value.

```python
# fill missing quantity using median
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```



* **Total Spent Column**

The missing values in the Total Spent column were handled by recalculating the total transaction amount using the Price Per Unit and Quantity columns. This ensures that the calculated values accurately represent the total purchase amount for each transaction.

```python
# recalculate total spent where it is missing
df['Total Spent'] = df['Price Per Unit'] * df['Quantity']
```



* **Discount Applied Column**

The Discount Applied column contains boolean values (True/False), where True indicates a discount was applied and False indicates no discount. Missing values in this column were replaced with False, assuming that no discount was applied when the value is missing. This ensures data consistency and maintains the correct data type.

```python
# Fill missing values in Discount Applied column with "No"
df['Discount Applied'] = df['Discount Applied'].fillna(False)
```

```python
# check if there is missing values
df.isnull().sum()
```
